## plot velocity profiles across a phase

```python
%matplotlib inline 
%config InlineBackend.figure_format = 'svg'
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tt = '../postProcessing/probesDict/0/B'
tt1= '/run/media/bini/BINIPhD/GaetanoBini/ripple/ks+1Hb1e4/postProcessing/probesDict/0/B'
```

```python
# probesDict for KOSST, ks+1 and Yw = 5e04

data_1 =np.loadtxt(tt)
```

```python
# probesDict for KOSST, ks+1 and Yw = 1e04
data_2 =np.loadtxt(tt1)
```

```python
# Data collected from Du Toit paper figure 15
data_3 =np.loadtxt('UExp_phase_crest')
data_4 =np.loadtxt('UExp_phase_crest_BL')
data_5 =np.loadtxt('UExp_phase_crest_SL')
data_6 =np.loadtxt('UExp_phase_trough')
data_7 =np.loadtxt('UExp_phase_trough_BL')
data_8 =np.loadtxt('UExp_phase_trough_SL')
```

```python
T = 5.37
pi = np.pi
a = 2*pi/T
Uo = 0.14
b = 1/Uo
c = pi/180
```

```python
plt.rcParams['figure.figsize'] = (8,7)
plt.rcParams.update({'font.size':17})

plt.plot ((data_2[:,0])*a - 3*pi,-(data_2[:,7])*b, label= r'$Y_w = 1 x 10^{-4}$')
plt.plot ((data_1[:,0])*a - 3*pi,-(data_1[:,7])*b, label= r'$Y_w = 5 x 10^{-4}$')

#plt.plot((data_3[:,0])*c,data_3[:,1],'o', ms=5, label='Sleath')
plt.plot((data_5[:,0])*c,data_5[:,1],'-.', ms=5, label='Traverse A')
plt.plot((data_4[:,0])*c,data_4[:,1],'-.', ms=5, label='Traverse B')

plt.xticks([0, pi/2, pi, 3*pi/2, 2*pi],
          ['$0$','$\dfrac{\pi}{2}$', '$\pi$','$\dfrac{3 \pi}{2}$', '$2\pi$'])
plt.title("$U_x$ over ripple crest")
plt.xlabel("$\Phi$")         
plt.ylabel('$\dfrac{U}{U_0}$')
plt.axhline(0, color= 'black', lw=1)
plt.ylim(-2.4,2.4)
plt.xlim(0,2*pi)
plt.legend (loc='best',fontsize=10)
plt.grid()
plt.savefig("crest.png")
```

```python
plt.rcParams['figure.figsize'] = (10,7)
plt.rcParams.update({'font.size':17})
#plt.plot((data_6[:,0])*c,data_6[:,1],'o', ms=5, label='Sleath')
plt.plot ((data_1[:,0])*a - (5*pi/2)+ 2*pi,-(data_1[:,4])*b, label= r'$Y_w = 5 x 10^{-4}$')
plt.plot ((data_2[:,0])*a - (5*pi/2)+ 2*pi,-(data_2[:,4])*b, label= r'$Y_w = 1 x 10^{-4}$')

#plt.plot((data_8[:,0])*c,data_8[:,1],'-.', ms=5, label='Traverse A')
#plt.plot((data_7[:,0])*c,data_7[:,1],'-.', ms=5, label='Traverse B')

plt.xticks([0, pi, 2*pi, 3*pi, 4*pi, 5*pi, 6*pi],
           ['$0$','$\pi$', '$2\pi$','$3\pi$', '$4\pi$','$5\pi$','$6\pi$'])
plt.title("$U_x$ over ripple trough")
plt.xlabel("$\Phi$")         
plt.ylabel('$\dfrac{U}{U_0}$')
plt.axhline(0, color= 'black', lw=1)
plt.ylim(-1.2,1.2)
plt.xlim(0,6*pi)
plt.legend (loc='best',fontsize=10)
plt.grid()
plt.savefig("trough.png")
```

### plot F_X/L  over time 

```python
data_1 =np.loadtxt('/home/bini/OpenFOAM/run/ripple/turbulent/data/F_X_Y_time260.csv', delimiter=',',skiprows=1)
```

```python
F_Y  =  data_1[:,272] 
F_X  =  data_1[:,265] 
time =  data_1[:,303] 
L = 0.17 
ss = pd.Series(F_X)
F_XX = ss.cumsum()
yy = pd.Series(F_Y)
F_YY = yy.cumsum()
```

```python
plt.rcParams['figure.figsize'] = (6,6)
plt.rcParams.update({'font.size':17})

plt.plot ((time)*a - 8*pi, F_X/L, label= r'$\dfrac{F_X}{L}$')
plt.plot ((time)*a - 8*pi, F_Y/L, label= r'$\dfrac{F_Y}{L}$')
plt.xticks([0, pi/2, pi, 3*pi/2, 2*pi],
          ['$0$','$\dfrac{\pi}{2}$', '$\pi$','$\dfrac{3 \pi}{2}$', '$2\pi$'])
# plt.title("$\dfrac{F_X}{L}$ and $\dfrac{F_Y}{L}$")
plt.xlabel("$\Phi$")         
plt.ylabel('$\dfrac{F}{L}$')
plt.axhline(0, color= 'black', lw=1)
#plt.ylim(-0.005,0.005)
plt.xlim(0,2*pi)
plt.legend (loc='best',fontsize=10)
plt.grid()
plt.savefig("Forces.png")
```

## cummulative $F_X$ plot over phase

```python
plt.rcParams['figure.figsize'] = (6,6)
plt.rcParams.update({'font.size':17})

plt.plot(time*a, F_XX/L, 'k--', linewidth=1.5,label= r'$\dfrac{F_{X,cum}}{L}$')
plt.plot(time*a, F_YY/L, 'g--', linewidth=1.5,label= r'$\dfrac{F_{Y,cum}}{L}$')
plt.xticks([0, pi, 2*pi, 3*pi, 4*pi, 5*pi, 6*pi, 7*pi, 8*pi, 9*pi, 10*pi],
           ['$0$','$\pi$', '$2\pi$','$3\pi$', '$4\pi$','$5\pi$','$6\pi$','$7\pi$','$8\pi$','$9\pi$','$10\pi$' ])

plt.xlabel("$\Phi$")         
plt.ylabel('$\dfrac{F}{L}$')
plt.axhline(0, color= 'black', lw=1)
#plt.ylim(-0.005,0.005)
plt.xlim(0,10*pi)
plt.legend (loc='best',fontsize=10)
plt.grid()
plt.savefig("cumm_Forces.png")
```
